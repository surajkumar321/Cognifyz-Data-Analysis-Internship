```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
df.shape
```

```python
df.head()
```

```python
price_count = df["Price range"].value_counts().sort_index()

price_count
```

```python
# Calculate Percentage

price_percentage = ((price_count / len(df)) * 100).round(2)

price_percentage
```

```python
plt.figure(figsize=(10,6))

ax = price_count.plot(
    kind="bar",
    color=["#4CAF50", "#2196F3", "#FF9800", "#E91E63"],
    edgecolor="black",
    linewidth=1.2
)

plt.title(
    "Distribution of Restaurants by Price Range",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.xlabel("Price Range", fontsize=12, fontweight="bold")
plt.ylabel("Number of Restaurants", fontsize=12, fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.4)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xticks(rotation=0)

# Data Labels
for bar in ax.patches:
    ax.annotate(
        f"{int(bar.get_height())}",
        (bar.get_x() + bar.get_width()/2, bar.get_height()),
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

plt.tight_layout()
plt.show()
```

```python
price_df = pd.DataFrame({
    "Price Range": price_percentage.index,
    "Percentage (%)": price_percentage.values
})

price_df
```

## Insights

- Price Range 1 has the highest number of restaurants.
- Price Range 2 is the second most common category.
- As the price range increases, the number of restaurants generally decreases.
- High-price restaurants are fewer compared to budget-friendly restaurants.

## Conclusion

The analysis shows that most restaurants belong to the lower price range categories, indicating that affordable dining options dominate the dataset. Premium-priced restaurants make up a smaller share.
