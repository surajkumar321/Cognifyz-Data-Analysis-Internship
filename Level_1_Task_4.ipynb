```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
online_delivery = df["Has Online delivery"].value_counts()

online_delivery
```

```python
online_percentage = (
    (online_delivery / len(df)) * 100
).round(2)

online_percentage
```

```python
# Professional Bar Chart - Online Delivery Availability

plt.figure(figsize=(8,5))

ax = online_delivery.plot(
    kind="bar",
    color=["#E74C3C", "#2ECC71"],
    edgecolor="black",
    linewidth=1.2
)

plt.title(
    "Restaurants Offering Online Delivery",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.xlabel("Online Delivery", fontsize=12, fontweight="bold")
plt.ylabel("Number of Restaurants", fontsize=12, fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.4)

# Remove Top & Right Border
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xticks(rotation=0)

# Data Labels
for bar in ax.patches:
    ax.annotate(
        f"{int(bar.get_height())}",
        (bar.get_x() + bar.get_width()/2, bar.get_height()),
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

plt.tight_layout()
plt.show()
```

```python
rating_comparison = (
    df.groupby("Has Online delivery")["Aggregate rating"]
    .mean()
    .round(2)
)

rating_comparison
```

```python
plt.figure(figsize=(8,5))

ax = rating_comparison.plot(
    kind="bar",
    color=["#3498DB", "#F39C12"],
    edgecolor="black",
    linewidth=1.2
)

plt.title(
    "Average Restaurant Rating by Online Delivery",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.xlabel("Online Delivery", fontsize=12, fontweight="bold")
plt.ylabel("Average Rating", fontsize=12, fontweight="bold")

plt.grid(axis="y", linestyle="--", alpha=0.4)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xticks(rotation=0)

# Data Labels
for bar in ax.patches:
    ax.annotate(
        f"{bar.get_height():.2f}",
        (bar.get_x() + bar.get_width()/2, bar.get_height()),
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

plt.tight_layout()
plt.show()
```

## Insights

- 74.34% of restaurants do not provide online delivery.
- Only 25.66% of restaurants offer online delivery.
- Restaurants with online delivery generally have higher average ratings.
- Online delivery services are available in a smaller portion of the restaurants in the dataset.

## Conclusion

The analysis shows that most restaurants do not provide online delivery services. However, restaurants offering online delivery tend to have better average ratings, indicating a possible relationship between online service availability and customer satisfaction.
