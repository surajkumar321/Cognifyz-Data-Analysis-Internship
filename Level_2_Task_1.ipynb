```python
# Distribution of Aggregate Ratings

rating_count = (
    df["Aggregate rating"]
    .value_counts()
    .sort_index()
)

rating_count
```

```python
# Professional Bar Chart - Aggregate Rating Distribution

plt.figure(figsize=(14,6))

ax = rating_count.plot(
    kind="bar",
    color="#4E79A7",
    edgecolor="black",
    linewidth=1.2
)

plt.title(
    "Distribution of Aggregate Ratings",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.xlabel("Aggregate Rating", fontsize=12, fontweight="bold")
plt.ylabel("Number of Restaurants", fontsize=12, fontweight="bold")

plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Remove Top & Right Border
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Data Labels
for bar in ax.patches:
    ax.annotate(
        f"{int(bar.get_height())}",
        (bar.get_x()+bar.get_width()/2, bar.get_height()),
        ha="center",
        va="bottom",
        fontsize=7
    )

plt.tight_layout()
plt.show()
```

```python
# Calculate Average Votes

average_votes = df["Votes"].mean()

print(f"Average Number of Votes: {average_votes:.2f}")
```

```python
# Create Rating Ranges

rating_bins = [0, 2, 3, 4, 5]

rating_labels = [
    "Poor (0-2)",
    "Average (2-3)",
    "Good (3-4)",
    "Excellent (4-5)"
]

df["Rating Range"] = pd.cut(
    df["Aggregate rating"],
    bins=rating_bins,
    labels=rating_labels,
    include_lowest=True
)

rating_range = df["Rating Range"].value_counts()

rating_range
```

```python
print("Most Common Rating Range:", rating_range.idxmax())
print("Number of Restaurants:", rating_range.max())
```

```python
# Rating Range Distribution

plt.figure(figsize=(8,5))

ax = rating_range.plot(
    kind="bar",
    color=["#E74C3C", "#F39C12", "#3498DB", "#2ECC71"],
    edgecolor="black",
    linewidth=1.2
)

plt.title(
    "Distribution of Rating Ranges",
    fontsize=15,
    fontweight="bold"
)

plt.xlabel("Rating Range", fontsize=12, fontweight="bold")
plt.ylabel("Number of Restaurants", fontsize=12, fontweight="bold")

plt.xticks(rotation=0)

plt.grid(axis="y", linestyle="--", alpha=0.3)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Data Labels
for bar in ax.patches:
    ax.annotate(
        f"{int(bar.get_height())}",
        (bar.get_x()+bar.get_width()/2, bar.get_height()),
        ha="center",
        va="bottom",
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

## Insights

- Most restaurants fall under the **Good (3–4)** rating category with **4,388 restaurants**.
- Around **2,158 restaurants** are in the **Poor (0–2)** category, mainly due to unrated (0.0) restaurants.
- Restaurants with **Excellent (4–5)** ratings are comparatively fewer (**1,114**).
- The average number of customer votes per restaurant is **156.91**, indicating moderate customer engagement.

## Conclusion

The analysis indicates that the majority of restaurants have **Good (3–4)** aggregate ratings. A considerable number of restaurants are unrated or have low ratings, while only a smaller portion achieve excellent ratings. On average, restaurants receive **156.91 customer votes**, reflecting moderate customer interaction.
